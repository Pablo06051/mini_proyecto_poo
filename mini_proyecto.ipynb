```python
class alumno:
    def __init__(self, cedula, nombre, edad, nota):
        self.__cedula = cedula 
        self.__nombre = nombre
        self.__edad = edad
        self.__nota = nota

    def get_cedula(self):
        return self.__cedula
    def get_nombre(self):
        return self.__nombre
    def get_edad(self):
        return self.__edad
    def get_nota(self):
        return self.__nota

    def set_nombre(self, nombre):
        self.__nombre = nombre
    def set_edad(self, edad):
        self.__edad = edad
    def set_nota(self, nota):
        self.__nota = nota
        if 0 <= nota <= 10:
            self.__nota = nota
        else:
            print("Advertencia: La nota debe estar entre 0 y 10.")

    def puede_votar(self):
        return self.__edad >= 18
         

    def aprueba_materia(self, nota_aprobacion=6.0): 
     return self.__nota >= nota_aprobacion
    def __str__(self):
        return (f"Cedula: {self.__cedula}, nombre: {self.__nombre},"
                f"Edad: {self.__edad}, nota: {self.__nota}")


alumno1 = alumno("12345", "Juan Perez", 20, 7.5)
alumno2 = alumno("67890", "Maria Lopez", 17, 8.0)



print(alumno1)
print(alumno2)

print(f"{alumno1.get_nombre()} puede votar: {alumno1.puede_votar()}")
print(f"{alumno2.get_nombre()} puede votar: {alumno2.puede_votar()}")


alumno1.set_nota(9.5)
print(f"{alumno1.get_nombre()} nueva nota: {alumno1.get_nota()}")
alumno1.set_nota(11.0)  # Intento de establecer una nota inválida
print(f"{alumno1.get_nombre()} nota después del intento inválido: {alumno1.get_nota()}")


```
